```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.worldometers.info/coronavirus/'
response = get(url)
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Table 
casos = html_soup.find_all('tr')
```

```python
value=[]
for i in range(0,len(casos)):
  l=casos[i].get_text().split('\n')[1:11]
  if l[0]=='Country,Other':
    value.append(i)
```

```python
col= ['Country','TotalCases','NewCases','TotalDeaths','NewDeaths','TotalRecovered','ActiveCases','Serious_Critical','TotCases1Mpop','TotDeaths1Mpop']
df=pd.DataFrame()

Country=[]
TotalCases=[]
NewCases=[]
TotalDeaths=[]
NewDeaths=[]
TotalRecovered=[]
ActiveCases=[]
Serious_Critical=[]
TotCases1Mpop=[]
TotDeaths1Mpop=[]

for i in range(1,value[1]):
  l=casos[i].get_text().split('\n')[1:11]
  Country.append(l[0])
  TotalCases.append(l[1].replace(',',''))
  NewCases.append(l[2].replace(',',''))
  TotalDeaths.append(l[3].replace(',',''))
  NewDeaths.append(l[4].replace(',',''))
  TotalRecovered.append(l[5].replace(',',''))
  ActiveCases.append(l[6].replace(',',''))
  Serious_Critical.append(l[7].replace(',',''))
  TotCases1Mpop.append(l[8].replace(',',''))
  TotDeaths1Mpop.append(l[9].replace(',',''))
  

df=pd.DataFrame(zip(Country,TotalCases,NewCases,TotalDeaths,NewDeaths,TotalRecovered,
                    ActiveCases,Serious_Critical,TotCases1Mpop,TotDeaths1Mpop),columns=col)

df
```

```python
df.to_csv('coronavirus.csv',header=True,index=False)
```
